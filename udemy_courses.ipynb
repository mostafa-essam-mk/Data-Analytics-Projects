## Dataset: https://www.kaggle.com/datasets/andrewmvd/udemy-courses

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('udemy_courses.csv', parse_dates=['published_timestamp'])
```

# [6] Check For Duplicate Data & Drop Them

```python
print('Is There any Duplicated Values?', df.duplicated().any())
```

```python
df.drop_duplicates(inplace=True)
```

# [7] Find Out Number of Courses Per Subjects

```python
df['subject'].value_counts()
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.countplot(data=df, x='subject', order=df['subject'].value_counts().index)
plt.xlabel('Subjects')
plt.ylabel('No. of courses per subject')
plt.show()
```

# [8] For Which Levels, Udemy Courses Providing The Courses

```python
df['level'].unique()
```

```python
df['level'].value_counts()
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.countplot(data=df, x='level')
plt.xlabel('Levels')
plt.ylabel('No. of courses per level')
plt.show()
```

# [9] Display The Count of Paid & Free Courses

```python
df['is_paid'].value_counts()
```

```python
plt.figure(figsize=(3, 3), dpi=150)
plt.pie(x=df['is_paid'].value_counts(), labels=['Paid', 'Free'], autopct='%1.0f%%')
plt.title('Paid Vs Free Courses')
plt.show()
```

# [10] Which Cours Has More Lectures (Free or Paid)?

```python
df.groupby('is_paid').mean(numeric_only=True)
```

# [11] Which Courses Have a Higher Number of Subscribers (Free or Paid)?

```python
df.groupby('is_paid').mean(numeric_only=True)
```

```python
sns.barplot(data=df, x='is_paid', y='num_subscribers')
plt.ylabel('No. of Subscribers')
plt.show()
```

# [12] Which Level has The Highest Number of Subscribers?

```python
df.groupby('level')['num_subscribers'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 5), dpi=100)
sns.barplot(data=df, x='level', y='num_subscribers')
plt.ylabel('No. of Subscribers')
plt.show()
```

# [13] Find The Most Popular Course Title

```python
df.nlargest(1, 'num_subscribers')['course_title']
```

# [14] Display 10 Most Popular Courses as per Number of Subscribers

```python
top_10 = df.nlargest(10, 'num_subscribers')
```

```python
plt.figure(figsize=(8, 6), dpi=200)
sns.barplot(data=top_10, y='course_title', x='num_subscribers')
# plt.xticks(rotation=90)
plt.show()
```

# [15] Find The Course Subject that having The Highest Number of Reviews 

```python
df.groupby('subject')['num_reviews'].sum(numeric_only=True).sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 5), dpi=100)
sns.barplot(data=df, x='subject', y='num_reviews')
plt.ylabel('No. of Reviews')
```

# [16] Does Price Affect Number of Reviews?

```python
plt.figure(figsize=(15, 6), dpi=200)
sns.scatterplot(data=df, x='price', y='num_reviews');
```

# [17] Find Total Number of Courses Related to Python

```python
df['course_title'].str.contains('Python', case=False).sum()
```

# [18] Display 10 Most Popular Python Courses as per number of Subscribers

```python
python_courses = df[ df['course_title'].str.contains('Python', case=False) ]
python_courses.nlargest(10, 'num_subscribers')
```

```python
plt.figure(figsize=(8, 5), dpi=200)
sns.barplot(data=python_courses.nlargest(10, 'num_subscribers'), y='course_title', x='num_subscribers');
plt.xlabel('No. of Subscribers')
plt.ylabel('Course Title')
plt.show()
```

# [19] In Which Year The Highest Number of Courses Were Posted?

```python
df['Year'] = df['published_timestamp'].dt.year
```

```python
df['Year'].value_counts().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 5), dpi=100)
sns.countplot(data=df, x='Year')
plt.show()
```

# [20] Display category-wise count of posted Subjects [year-wise]

```python
df.groupby(['Year', 'subject']).count()
```

```python
plt.figure(figsize=(12, 6), dpi=150)
sns.countplot(data=df, x='Year', hue='subject')
plt.show()
```
